```python
ds = pd.read_csv('confirmed_detailed_list.csv',low_memory='False')
```

```python
pd.set_option('max_columns',None)
pd.set_option('max_rows',None)
ds.head(150)
```

# Removing NULL Values

```python
print("Number of objects: "+str(len(ds.index)))
ds.dropna(inplace=True)
ds.isnull().sum()
```

# Selecting Dependent & Independent Variables

```python
x1=ds.iloc[ : , 2:-1].values
y1=ds.iloc[ : , -1].values
```

```python
print(y1)
```

```python
print(x1)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x1,y1,test_size = 0.3,random_state=0)
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression(solver='lbfgs',class_weight='balanced', max_iter=1000)
```

```python
logmodel.fit(x_train,y_train)
```

```python
prediction=logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,prediction)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cf_matrix=confusion_matrix(y_test,prediction)
print(cf_matrix)
```

```python
sb.heatmap(cf_matrix, annot=True)
```
